# CREATION D'UN ENVIRONNEMENT VIRTUEL

# A. Sous Windows

Sur Windows, appelez la commande venv comme suit :

### 1. Via le prompt CMD (Incite de commande Windows)

``python -m venv c:\path\to\myenv``

Pour activer l'environnement :

- Se place dans le répertoire : ``myenv``  

- Saisir : ``.\Scripts\activate``  

Pour désactiver l'environnement, saisir : ``deactivate``

### 2. Via Anaconda prompt

Création : 
- ``conda create -n myenv``  

Activation :
- ``conda activate myenv``  



# B. Sous Linux/ MacOS

Création :

``python3 -m venv /path/to/new_env``

Activation :
Pour activer l'environnement, Saisir :
``source path/to/new_env/bin/activate``

```python
import requests
from bs4 import BeautifulSoup
```

```python
# Fait une demande au server
URL = "https://www.jumia.ci/smartphones/"
# URL = "https://www.abidjan.net/sante/pharmacies/"
page = requests.get(URL)
page
```

- Ouvrir un navigateur Firefox  
- Charger la page d'accueil de  Yahoo  


```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
# Le chemin du driver
gecko_driver_path = "../PRATIQUES_WEB_SCRAPING/scraping_env/geckodriver.exe"
# Ouvrir le navigateur
browser = webdriver.Firefox(executable_path=gecko_driver_path)
# Charger la page
browser.get('http://www.yahoo.com')
```

Obtenir le titre de la page

```python
browser.title
```

```python
browser.title
```

- rechercher "seleniumhq" 


```python
assert('Yahoo' in browser.title)

# elem = browser.find_element_by_name('p')  # Find the search box
elem = browser.find_element(By.NAME, 'p') 
elem.send_keys('seleniumhq' + Keys.RETURN)
```

- fermer le navigateur  

```python
browser.quit()
```

#### Chrome

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
# Le chemin du driver
gecko_driver_path = "../WEB_DRIVERS/chromedriver.exe"
# Ouvrir le navigateur
browser = webdriver.Chrome(executable_path=gecko_driver_path)
# Charger la page
browser.get('http://www.yahoo.com')
```

```python
browser.title
```

```python
browser.title
```

```python
assert('Yahoo' in browser.title)

elem = browser.find_element(By., 'ybar-sbq')  # Find the search box
elem.send_keys('seleniumhq' + Keys.RETURN)
```
